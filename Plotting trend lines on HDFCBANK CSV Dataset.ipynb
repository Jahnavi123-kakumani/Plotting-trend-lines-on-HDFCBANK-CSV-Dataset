<p style="color:#FFFFFF; font-size:30px; background-color:#007BFF; padding:10px 20px; display:inline-block; border-radius:5px; text-align:center; cursor:pointer;">Step 5: Identify the Trend Lines</p>


```python
# Simple example to plot a trend line - this would need more complex logic for robust trend detection
# For demonstration, assume a basic trend line for an uptrend and downtrend manually.

# Example uptrend and downtrend points (can be computed more dynamically)
uptrend_x = pd.to_datetime(['2023-01-01', '2023-03-01', '2023-05-01'])
uptrend_y = [1500, 1550, 1600]

downtrend_x = pd.to_datetime(['2023-01-01', '2023-03-01', '2023-05-01'])
downtrend_y = [1650, 1600, 1550]

plt.figure(figsize=(10, 6))
plt.plot(df['Close'], label='HDFCBANK Closing Price', color='blue')

# Plot uptrend and downtrend lines (example)
plt.plot(uptrend_x, uptrend_y, label='Uptrend', color='green', linestyle='--')
plt.plot(downtrend_x, downtrend_y, label='Downtrend', color='red', linestyle='--')

plt.title('HDFCBANK Trend Lines')
plt.xlabel('Date')
plt.ylabel('Closing Price (INR)')
plt.grid(True)
plt.legend()
plt.show()
```

<p style="color:#FFFFFF; font-size:30px; background-color:#007BFF; padding:10px 20px; display:inline-block; border-radius:5px; text-align:center; cursor:pointer;">Step 6: Annotate the Trend Lines</p>


```python
plt.figure(figsize=(10, 6))
plt.plot(df['Close'], label='HDFCBANK Closing Price', color='blue')

# Plot trend lines
plt.plot(uptrend_x, uptrend_y, label='Uptrend', color='green', linestyle='--')
plt.plot(downtrend_x, downtrend_y, label='Downtrend', color='red', linestyle='--')

# Annotate trend lines
plt.annotate('Uptrend', xy=(uptrend_x[1], uptrend_y[1]), xytext=(uptrend_x[1], uptrend_y[1] + 10),
             arrowprops=dict(arrowstyle="->", lw=1), fontsize=12, color='green')
plt.annotate('Downtrend', xy=(downtrend_x[1], downtrend_y[1]), xytext=(downtrend_x[1], downtrend_y[1] - 10),
             arrowprops=dict(arrowstyle="->", lw=1), fontsize=12, color='red')

plt.title('HDFCBANK Trend Lines')
plt.xlabel('Date')
plt.ylabel('Closing Price (INR)')
plt.grid(True)
plt.legend()
plt.show()
```

<body>
    <h2>Insights for Trend Lines Plotting Code</h2>
    <ul>
        <li><strong>Setting the Figure Size:</strong> The figure size is set to (10, 6) to ensure that the plot is clear and easy to interpret. A larger plot is often needed for a better visual representation of trends.</li>
        <li><strong>Plotting the Closing Price:</strong> The closing price of HDFCBANK is plotted over time to track the stock's performance. The label 'HDFCBANK Closing Price' helps identify the main data on the plot.</li>
        <li><strong>Drawing Trend Lines:</strong> Trend lines are added to indicate the uptrend and downtrend using dashed lines. The 'uptrend' line is green, and the 'downtrend' line is red to easily differentiate the two trends.</li>
        <li><strong>Annotating the Trend Lines:</strong> The `annotate()` function is used to label the trend lines with 'Uptrend' and 'Downtrend' at key points. The annotations are positioned using the coordinates of the trend points, and arrows are used to direct attention to the trend lines.</li>
        <li><strong>Chart Customization:</strong> The title, labels, grid, and legend are added for clarity. The title 'HDFCBANK Trend Lines' provides context, while the x-axis and y-axis labels explain the data being presented.</li>
        <li><strong>Visual Representation:</strong> The use of different colors (green for uptrend and red for downtrend) allows the trends to be easily recognized. The plot grid enhances readability, especially when tracking price changes over time.</li>
        <li><strong>Final Display:</strong> The `plt.show()` function displays the plot, allowing the user to visually analyze the HDFCBANK stock's trends over time, highlighting significant periods of growth and decline.</li>
    </ul>

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Close'], label='HDFCBANK Closing Price', color='blue')

# Plot trend lines
plt.plot(uptrend_x, uptrend_y, label='Uptrend', color='green', linestyle='--')
plt.plot(downtrend_x, downtrend_y, label='Downtrend', color='red', linestyle='--')

# Annotate trend lines
plt.annotate('Uptrend', xy=(uptrend_x[1], uptrend_y[1]), xytext=(uptrend_x[1], uptrend_y[1] + 10),
             arrowprops=dict(arrowstyle="->", lw=1), fontsize=12, color='green')
plt.annotate('Downtrend', xy=(downtrend_x[1], downtrend_y[1]), xytext=(downtrend_x[1], downtrend_y[1] - 10),
             arrowprops=dict(arrowstyle="->", lw=1), fontsize=12, color='red')

# Save the plot
plt.title('HDFCBANK Trend Lines')
plt.xlabel('Date')
plt.ylabel('Closing Price (INR)')
plt.grid(True)
plt.legend()
plt.savefig('hdfcbank_trend_lines.png')
```
